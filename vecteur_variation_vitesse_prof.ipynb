# Vecteur variation de vitesse

## Document 1 : Mthode de trac d'un vecteur vitesse $\overrightarrow{v}$
On a reprsent ci-dessous diffrentes positions dune balle  intervalles de temps courts et gaux au cours de son mouvement.



Le vecteur vitesse   de la balle  la position $M_4$ est considr comme gal au vecteur vitesse moyenne entre les positions $M_3$ et $M_5$.
On calcule la valeur du vecteur vitesse moyenne entre les positions $M_3$ et $M_5$:
$$v_4=\frac{M_3M_5}{t_5-t_3} $$

$t_3$ et $t_5$ sont les dates proches de passage en $M_3$ et $M_5$.
En fonction de la valeur $v_4$ de la vitesse, on propose une chelle.

On trace le vecteur vitesse $\overrightarrow{v_4}$  partir de la position $M_4$, tangent  la trajectoire, dans le sens du mouvement. La longueur du segment flch est proportionnelle  la valeur de la vitesse.

````python
VAx = (X[i] - X[i-2]) / (T[i] - T[i-2])
VAy = (Y[i] - Y[i-2]) / (T[i] - T[i-2])
VA = (VAx, VAy)
VBx = (X[i+2] - X[i]) / (T[i+2] - T[i])
VBy = (Y[i+2] - Y[i]) / (T[i+2] - T[i])
VB = (VBx, VBy)
````
